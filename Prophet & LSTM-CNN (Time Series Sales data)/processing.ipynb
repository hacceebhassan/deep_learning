```
### ----- Importing Libraries ----- ###
import numpy as np
import pandas as pd
import itertools
```

```
# Dataset
data = pd.read_excel('sales_data.xlsx')
```

### **Processing Raw Data**

```
# Getting Company Names
companies = data['Company Name'].tolist()

# Dropping column
data_m = data.drop(['Company Name'], axis=1)

# Making transpose
data_m_t = data_m.T
data_m_t.reset_index(drop=True, inplace=True)

# Getting dates
date = data.columns.tolist()
date.pop(0)

# Getting values
series = []
for i in range(610):
  series.extend(data_m_t[i].tolist())
  i += 1

# Repicating dates as per the data
date_ = list(itertools.repeat(date, int(len(series)/len(date))))
date_final = []
for dat in date_:
  for d in dat:
    date_final.append(d)

# Replicating companies as per the dataset
Comp = []
for c in companies:
  ass = list(itertools.repeat(c, int(len(date))))
  Comp.extend(ass)
```

```
# Validatinb lengths of our lists
print('Length of Companies : ', len(Comp))
print('Length of Dates : ', len(date_final))
print('Length of Values : ', len(series))
```

```
# Making a dataframe
df_final = pd.DataFrame(list(zip(Comp, date_final, series)), columns = ['company', 'date', 'sales'])
print(df_final)

# Appending dataframe to excel
df_final.to_excel('data/processed_sales_data.xlsx', index=False)
```
